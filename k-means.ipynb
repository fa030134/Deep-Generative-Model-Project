```python
from langchain_community.document_loaders import DirectoryLoader, TextLoader, PyMuPDFLoader, CSVLoader

from py2neo import Graph, Node, Relationship
from py2neo import NodeMatcher, RelationshipMatcher
import spacy
from tqdm import tqdm
import re

from sentence_transformers import SentenceTransformer

model_path="/home/yuxin.yang/model/moka-ai/m3e-base"
model = SentenceTransformer(model_path)


graph=Graph('http://localhost:7474',user='neo4j',password='cecaceca512',name='neo4j')
node_matcher = NodeMatcher(graph)

query = """
    MATCH (n)
    RETURN n.text AS text, n.embedding AS embedding
    """
results=graph.run(query).data()


node_texts = []
embeddings = []

for record in results:
    #print(record)
    node_texts.append(record['text'])
    embeddings.append(record['embedding'])

```

```python
from sklearn.cluster import KMeans
import numpy as np

# 将嵌入数据转换为numpy数组，因为scikit-learn期望这样的输入格式
embeddings_array = np.array(embeddings)

# 假设我们想将数据聚成3类
kmeans = KMeans(n_clusters=3, random_state=0).fit(embeddings_array)

# 获取每个节点的聚类标签
cluster_labels = kmeans.labels_
```
